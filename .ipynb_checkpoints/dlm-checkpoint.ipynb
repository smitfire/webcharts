```python
import pandas as pd
from sklearn import preprocessing
```

```python
def cleandf(sht):
    df = pd.read_excel("xl_sheets/Demand Discrepancies (DLM)- DOD.xlsx", sheetname=sht, header=1)
    df.drop(df.columns[[5,6]], axis=1, inplace=True)
    return df[df["Inbound Discrepancy"].notnull()].dropna(axis=1, how="all").fillna(0)

psdf = cleandf("Pulsepoint")
adxdf = cleandf("AdX")
aoldf = cleandf("AOL")
cdf = cleandf("Conversant")
sdf = cleandf("Smaato")
pbdf = cleandf("Pubmatic")

frames = [psdf,sdf,cdf, adxdf, aoldf,pbdf]
psdf.head()
# for cdf in frames:
#     cdf.drop('Outbound Discrepancy', 1, inplace=True)
#     cdf.drop('Fill Rate', 1, inplace=True)
```

```python
f = pd.concat(frames, keys=['Pulsepoint', 'Smaato', 'Conversant', 'AdX', 'AOL','Pubmatic'])
dates = [f["Date"].min(), '2015-10-08', '2015-10-09','2015-10-10', f["Date"].max()]
ef = pd.DataFrame({"Date": dates}, index=[3,4,5,6,7])

mekeys=['Pulsepoint', 'Smaato', 'Conversant', 'AdX', 'AOL','Pubmatic']

for mk in mekeys:
#     ef.insert(1, mk, f.ix[mk]["Inbound Discrepancy"])
#     ef.insert(1, mk, f.ix[mk]["Outbound Discrepancy"])
    ef.insert(1, mk, f.ix[mk]["Revenue:"])
#     ef.insert(1, mk, f.ix[mk]["Fill Rate"])


ef = ef.fillna(0)
ef.head(20)
# f.head(40)
# ef.plot(x="Date")
ef.to_csv("revtimedata.csv")
```

```python
# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

ef.head(10)
for col in ef:
    if col != "Date":
        x = ef[col].values.astype(float)
        x_sc = min_max_scaler.fit_transform(x)
        ef[col] = x_sc
ef.head()
ef.to_csv("revtimedata_norm.csv")
```

```python
psdf = cleandf("Pulsepoint")
adxdf = cleandf("AdX")
aoldf = cleandf("AOL")
cdf = cleandf("Conversant")
sdf = cleandf("Smaato")
pbdf = cleandf("Pubmatic")

frames = [psdf,sdf,cdf, adxdf, aoldf,pbdf]
psdf.head()
# psdf.to_csv("pp.csv")
```

```python
for i, df in enumerate(frames):
    for col in df:
        if col != "Date":
            x = df[col].values.astype(float)
            x_sc = min_max_scaler.fit_transform(x)
            df[col] = x_sc
    df.to_csv(str(i) + "_partner_norm.csv")

psdf.head()
sdf.head()
# psdf.to_csv("pulse_norm.csv")
```
